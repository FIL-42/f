### PROBLEM 1

**Solve for eigenvalues and eigenvectors of a symmetric matrix**

1. Use numpy's $\tt linalg$ package to find the eigenvalues and eigenvectors of the ff. symmetrix matrix *(15 pts.)*:

$ \begin{equation}
{\bf A} =
\begin{pmatrix}
8 & -2 & -20 \\
-2 & -3 & -9 \\
-20 & -9 & -3
\end{pmatrix} \end{equation} $

2. Check your answer numerically; that is, check that $\bf AV = VD$, where the $N$ x $N$ matrix $\bf V$ has the eigenvectors as its columns and $\bf D$ is the diagonal matrix with the eigenvalues $\lambda_i$ as its diagonal entries. *(15 pts.)*

```python
from numpy import array
from numpy.linalg import eigh
import numpy as np
a=array([[8,-2,-20],[-2,-3,-9],[-20,-9,-3]], float)
x, v = eigh(a) #eigh is used to find eigenvalues and eigenvectors
print("Problem 1.1 Answer")
print("These are the eigenvalues:",x)
print("These are the eigenvectors:")
print(v)
av=np.matmul(a,v) #matrix mult of a and v
d=array([[x[0],0,0],[0,x[1],0],[0,0,x[2]]], float) #diagonal matrix
vd=np.matmul(v,d) #matrix mult of v and d
print("If AV and VD are not rounded off...")
print(av==vd) #boolean will show almost all entries are not equal.
#it is because values of av and vd are equal only up to 7 decimals due to approximation error.
#Thus av and ad are rounded off to 7 decimals.
for i in range(3):
  for j in range(3):
    av[i][j]=round(av[i][j],7) #av rounded off to 7 decimals
    vd[i][j]=round(vd[i][j],7) #vd rounded off to 7 decimals
print("Problem 1.2 Answer")
print("IF AV an VD are rounded off to 7 decimals")
print("Boolean shows that AV=VD")
print(av==vd)

```

**Solve a nonlinear equation using the relaxation method**

Consider the equation $x=1- e^{-cx}$, where $c$ is a known parameter and $x$ is unknown. This equation arises in a variety of situations, including the physics of contact processes, mathematical models of epidemics, and the theory of random graphs.

1. Write a program to solve this equation for $x$ using the relaxation method for the case $c = 2$. Calculate your solution to an accuracy of at least $10^{-6}$. *(10 pts.)*

2. Modify your program to calculate the solution for values of $c$ from 0 to 3 in steps of 0.01. *(10 pts.)*

3. Make a plot of $x$ as a function of $c$. Describe the different regimes that you see. *(10 pts.)*

```python
#This is Number 2.1
from math import exp
x=1.0 #arbitrary x
def f(x):
  return 1-exp(-2*x) #function
for k in range(20):
  x=f(x) #old f(x) is the new f(x)
  #looped for 20 times
x=round(x,6) #"accuracy at least 10^-6". It is also found to be accurate from loops >14
print("Method converges to a value: ",x)
```

```python
#This is Number 2.2 and 2.3
from math import exp
import numpy as np
def f(x,c):
  return 1-exp(-c*x) #function
xlist=[] #list for y-axis in graph
clist=[] #list for x-axis in graph
for j in range(0,301): #total of 300 iterations -> from (3-0)*0.01
  x=1.0 #function at x=1
  e=1.0 #arbitrary error value
  c=j/100 #0.01 step
  clist.append(c) #incremental value is added to x-axis
  while e>10**(-6): #while loop uses error to continue approximating x until it is accurate by 10^-6
    x=f(x,c) #function call
    e=x-f(x,c) #error is the value of new f(x) and old f(x) and while loop causes approximates error to 0, closing into x value.
  xlist.append(x) #data is added to y-axis
#plot code
import matplotlib.pyplot as plt
plt.plot(clist, xlist)
plt.title("Graph of Convergences as a function of cases 0 to 3")
plt.xlabel("cases")
plt.ylabel("Values of Convergences")
plt.show()


```

### PROBLEM 3

**Physics Application: Calculating the Lagrange point**

_Refer to Exercise 6.16 of the Newman text._ The $L_1$ Lagrange point is the point between the Earth and the Moon at which a satellite will orbit the Earth in perfect synchrony with the Moon, staying always in between the two.

Assuming circular orbits, and assuming that the Earth is much more massive than either the Moon or the satellite, the distance $r$ from the center of the Earth to the $L_1$ point satisfies
$$ \begin{equation}
\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r,
\end{equation} $$
where $R$ is the distance from the Earth to the Moon, $M$ and $m$ are the Earth and Moon masses, $G$ is Newton's gravitational constant, and $\omega$ is the angular velocity of both the Moon and the satellite.

Use Newton's method to solve for the distance r from the Earth to the $L_1$ point. Compute a solution accurate to at least four significant figures.

```python
G=6.674*10**(-11) #all constants and given
M=5.974*10**(24)
m=7.348*10**(22)
R=3.844*10**8
w=2.662*10**(-6)
def f(r): #function s.t. all items are put on LHS and RHS=0
  return ((G*M)/r**2)-((G*m)/(R-r)**2)-w**2*r
def df(r): #derivative
  return -2*((G*M)/r**3)+2*((G*m)/(R-r)**3)-w**2
r=1 #arbitrary r
while abs(f(r))>1e-15: #the loop iterates until f(r) is approximately 0 and it will output the root r.
  r=r-(f(r)/df(r)) #this is r' in the slides.
  print(r)
r=round(r,-5) #4 sig figs
print("The distance of the satellite from Earth is: ", r)
print("That is, the satellite would be quite close to the Moon so that it will orbit in perfect synchrony with the Earth and Moon.")
```
